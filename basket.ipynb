```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


df = pd.read_excel(r"C:\Users\Hazeeb\OneDrive\Desktop\inter\New folder\online_retail_II (1)_Updated.xlsx")
df
```

```python
df['Customer ID'] = df['Customer ID'].dropna()
df['Description'] = df['Description'].fillna("unknown")
df.isnull().sum()
```

```python
df['Description'] = df['Description'].str.strip() # remove spaces
df['Invoice']=df['Invoice'].astype('str') # convert into str
df=df[~df['Invoice'].str.contains('c')] #remove credit
df
```

```python
basket = (df.groupby(['Invoice', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Invoice'))

basket
```

```python
basket_sets = basket.apply(lambda row: row.map(lambda quantity: 1 if quantity > 0 else 0), axis=1)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.5)
```

```python
rules.tail(5)
```

```python
#conditions

filtered_rules = rules[(rules['lift'] >= 3) & (rules['confidence'] >= 0.3)]
print(filtered_rules)
```

commonly purchased together and recommend product bundles 

```python
sorted_rules = filtered_rules.sort_values(by=['lift'], ascending=False)
print(sorted_rules.head(10))
```
